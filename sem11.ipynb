f(x) =0.6x^3+5.5x^2+10x-5

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = 5

x = Symbol('x')

f_x = 0.6*(x**3) + 5.5*(x**2) + 10*(x**1) - 5
```

```python
f_x
```

## 1. Определяем корни

```python
print("корень №1:", solve(f_x, x, cubics= False)[0].n(4))
print("корень №2:", solve(f_x, x, cubics= False)[1].n(4))
print("корень №3:", solve(f_x, x, cubics= False)[2].n(4))
```

## 2. Интервалы на которых функция возрастает

```python
derivative = diff(f_x)
derivative
```

```python
solve(derivative > 0, x)
```

## 3. Интервалы на которых функция убывает

```python
solve(derivative < 0, x)
```

## 4. Построение графика

```python
g = plot(f_x, derivative, (x, -6, 3), title = 'совмещенный график', legend= True, xlabel= 'X', ylabel= 'Y')
```

## 5. Вычисляем вершины

```python
max = solve(derivative, x)[0].n(3)
min = solve(derivative, x)[1].n(3)
print('Точки экстремума:')
print(f'Максимум = {max}, Минимум = {min}')
```

## 6. промежутки на котором f > 0

```python
solve(f_x > 0, x)
```

## 7. промежутки на котором f < 0

```python
solve(f_x < 0, x)
```
